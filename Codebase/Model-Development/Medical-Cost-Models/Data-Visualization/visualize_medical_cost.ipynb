```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
try:
    df = pd.read_csv("../../../Data/Medical-Cost-Data/medical_cost.csv")
except:
    print(f"Data not found, make sure to run the medical_cost_preprocessing.ipynb file in its entirety to retrieve the data")
```

```python
# Let's take a closer look at the data to understand how it is made up
sns.set_theme(style="whitegrid", palette="muted")

plt.figure(figsize=(12, 8))
sns.histplot(df['charges'], kde=True, bins=50)
plt.title('Distribution of Medical Charges', fontsize=16)
plt.xlabel('Charges by Dollars', fontsize=12)
plt.ylabel('Number of Patients', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(np.log1p(df['charges']), kde=True, bins=50, color='blue')
plt.title('Log Computed for Distribution of Medical Charges', fontsize=16)
plt.xlabel('Log(Charges by Dollars)', fontsize=12)
plt.ylabel('Number of Patients', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
features_of_interest = [
    'charges', 'age', 'sex', 'bmi', 'children', 'smoker', 'is_obese', 'senior', 'multi_children', 'age_squared', 'bmi_smoker', 'age_smoker'
]
correlation_matrix = df[features_of_interest].corr()

sns.heatmap(
    correlation_matrix, 
    annot=True, 
    fmt=".2f", 
    cmap='coolwarm', 
    linewidths=.5,
    annot_kws={"size": 10}
)

plt.title('Correlation Heatmap of Dataset Features', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=df,
    x='age',
    y='charges',
    hue='smoker',
    palette={0: 'blue', 1: 'red'},
    alpha=0.7,
    size='bmi',
    sizes=(20, 200)
)

plt.title('Medical Charges by Age, BMI, and Smoker Status', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Charges ($)', fontsize=12)
plt.legend(title='Smoker, 0=No & 1=Yes', loc='upper left')
plt.show()
```
